```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


if 'Freq [Hz]' in df_concatenated.columns:
    df_concatenated = df_concatenated.drop(columns=['Freq [Hz]'])

categorias = ['[-300, -250)', '[-250, -200)', '[-200, -150)', '[-150, -100)', '[-100, -50)', '[-50, 0)']
bins = [-300, -250, -200, -150, -100, -50, 0]

cores = {
    'S': '#FF00FF',
    'R': '#0000FF',
    'T': '#00FFFF',
    'H1': '#FF0000',
    'H2': '#FFA07A',
    'K1': '#008000',
    'K2': '#90EE90'
}

for coluna in df_concatenated.columns:
    cor = cores.get(coluna.strip(), 'gray')
    plt.figure()
    plt.hist(df_concatenated[coluna], bins=bins, edgecolor='black', color=cor)
    plt.title(f'Histograma de {coluna}')
    plt.xlabel('Categorias [dB]')
    plt.ylabel('Contagem')
    plt.xticks(bins[:-1], categorias, rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
